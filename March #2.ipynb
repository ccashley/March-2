```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_feather("/Users/ashleychen/Downloads/output_mar (1).feather")
```

```python
df
```

```python
df.info 
```

```python
random_sample = df.sample(n = 1000, random_state = 42) # for a fixed random sample
```

```python
random_sample
```

```python
random_sample['candidate'] = random_sample['text'].str.extract(r'RT @(\w+)') 
```

```python
random_sample['candidate']
```

```python
#convert date string to pd date-time object
random_sample['date'] = pd.to_datetime(random_sample['date'], format = '%a %b %d %H:%M:%S +0000 %Y') 
#sort day of the week for each date
random_sample['weekindex'] = random_sample['date'].dt.isocalendar().week
```

```python
tweets_per_day = random_sample.groupby(['candidate', 'weekindex']).size().reset_index(name = 'tweet_count') #new df
```

```python
tweets_per_day
```

```python
plt.figure(figsize = (12, 8))
sns.set(style = "whitegrid")
sns.lineplot(x = 'weekindex', y = 'tweet_count', hue = 'candidate', data = tweets_per_day, marker = 'o') #indicate data points
plt.title('Number of Tweets per Candidate Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Tweets')
plt.show()
```
